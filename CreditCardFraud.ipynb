```python
from pycaret.classification import setup, compare_models, evaluate_model, predict_model, save_model, load_model
```

```python
x=df.loc[:, df.columns != 'fraud']
```

```python
y=df['fraud']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
X_train1=x_train.join(y_train)
```

```python
X_test1=x_test.join(y_test)
```

```python
X_train1
```

```python
X_train1['fraud'].value_counts()
```

```python
X_test1['fraud'].value_counts()
```

```python
modelSet = setup(X_train1,target='fraud')
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
predict_model(best_model, data=X_test1)
```

```python
best_model
```

```python
save_model(best_model, 'decisiontree')
```
